```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('MEANS OF PAYMENT').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**Most of the purchases were made with card.It is recommendation to get kaufland and Netto  to ease my purchases.They are the shops i made most purchases.**

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('SHOP').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**The above graph have shown how often purchases were made within different shops.The chart clearly shows most of my purchases were made in Kaufland and Netto with the third highest in Asia Might.Kaufland and Netto were highest because they are very close to my apartment.Asia Might came third because i bought african foodstuffs from the shop though it was very far from my apartment.**
