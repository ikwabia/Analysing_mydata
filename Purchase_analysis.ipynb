```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data['DAY'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='AMOUNT(Euro)', y='DAY', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

**Much payments were made on Saturdays, is very ideal to make my future budgets considering Saturdays are meant to make much payments.**

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('MEANS OF PAYMENT').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**Most of the purchases were made with card, and therefore  as a recommendation get kaufland card and Netto card to ease my purchases.They are the shops i made most purchases.**

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('SHOP').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

**The above graph have shown how often purchases were made within different shops.The chart clearly shows most of my purchases were made in Kaufland and Netto with the third highest in Asia Might.Kaufland and Netto were highest because they are very close to my apartment.Asia Might came third because i bought african foodstuffs from the shop though it was very fr from my apartment.**
